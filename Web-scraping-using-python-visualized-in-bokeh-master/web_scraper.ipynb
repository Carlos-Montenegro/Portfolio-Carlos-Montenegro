```python
#writing the csv file 'airbnb_1'
csv_file = open('airbnbpricesLima.csv', 'w', encoding='utf-8')
writer = csv.writer(csv_file)
writer.writerow(['url','title','host','location','room_type','bed','bath','feat_amenities','rate','reviews','price'])
page_urls = ['https://www.airbnb.com/s/Lima--Peru/homes?refinement_paths%5B%5D=%2Fh\
omes&place_id=ChIJxz7uGfbFBZERSi5FzLlsIBQ&source=mc_search_bar&search_type=sectio\
n_navigation&screen_size=large&checkin=2020-02-01&checkout=2020-02-08&adults=1&section_of\
fset=8&items_offset={}'.format(18 * x) for x in range(0, 18)]
all_page_urls = []
index = 1
for url in page_urls:
    print("Scraping page number" + str(index))
    #We specify the location of the url from where we are going to take the code 
    a = driver.find_elements_by_xpath('//div[@class="_8ssblpx"]/div/div/div/div/a')
    #We take the urls of the locations previously specified
    detail_urls = [x.get_attribute("href") for x in a]
    #number of urls for specific properties to scrape
    driver.get(url)
    print(str(len(detail_urls)) + " links are going to be scraped in this page")
    print('=' * 50)
    #We put the url in our list
    all_page_urls.extend(detail_urls)
    index += 1
index = 1 
print("Total of rentals: {}".format(len(all_page_urls)))
print('=' * 50)
```

```python
index =0
for url in all_page_urls:
    print("Scraping rental number {} - {:0.2f}% completed".format(index, (index / len(all_page_urls) * 100)))
    driver.get(url)
    result = {}

    driver.wait_rental = WebDriverWait(driver, 5)

    result['url']=url
    try:
        result['title'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="summary"]/div/div/div[1]/div/div/div[1]/div[1]/div/span/h1/span'))).text
    except:
        result['title'] = None   

    try:
        result['host'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="summary"]/div/div/div[1]/div/div/div[2]/div/div/div[2]'))).text
    except:
        result['host'] =None

    try:
        result['location'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="summary"]/div/div/div[1]/div/div/div[1]/div[2]/div/a/div'))).text
    except:
        result['location'] = None     

    try:
        result['room_type'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="room"]/div[2]/div/div[2]/div[1]/div/div[3]/div/div/div/div[1]/div/div/div[2]/div'))).text
    except:
        result['room_type'] = None     
   
    try:
        result['bed'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="room"]/div[2]/div/div[2]/div[1]/div/div[3]/div/div/div/div[1]/div/div/div[3]/div'))).text
    except:
        result['bed'] =None

    try:
        result['bath'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="room"]/div[2]/div/div[2]/div[1]/div/div[3]/div/div/div/div[1]/div/div/div[4]/div '))).text
    except:
        result['bath'] =None

    try:
        result['feat_amenities'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="amenities"]/div/div/div/div/div/section/div[5]/div/button'))).text
    except:
        try:
            result['feat_amenities'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="amenities"]/div/div/div/div/div/section/div[4]/div/button'))).text
        except:
            result['feat_amenities']=None

    try:
        result['rate'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="room"]/div[2]/div/div[2]/div[2]/div/div/div/div[2]/div/div/div[1]/div/div/div/div[1]/div/div/div[2]/button/div/div[1]/div[2]/div/div'))).text
    except:
        result['rate'] =None

    try:
        result['reviews'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="room"]/div[2]/div/div[2]/div[2]/div/div/div/div[2]/div/div/div[1]/div/div/div/div[1]/div/div/div[2]/button/div/div[2]/span'))).text
    except:
        result['reviews'] =None
   
    try:
        result['price'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="room"]/div[2]/div/div[2]/div[2]/div/div/div/div[2]/div/div/div[1]/div/div/div/div[1]/div/div/div[1]/div/span[2]/span'))).text
    except:
        try:
            result['price'] = driver.wait_rental.until(EC.presence_of_element_located((By.XPATH,'//*[@id="room"]/div[2]/div/div[2]/div[2]/div/div/div/div[2]/div/div/div[1]/div/div/div/div[1]/div/div/div[1]/div/span[3]/span'))).text
        except:
            result['price'] =None
        
    print(result)
        
    index += 1
    writer.writerow(result.values())


csv_file.close()
#driver.close()
```
